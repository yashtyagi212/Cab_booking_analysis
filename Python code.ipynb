```python
import pandas as pd

# Load data
df = pd.read_csv("uber-raw-data-apr14.csv")

# Check first 5 rows
print(df.head())

# Check dataset info
print(df.info())

# Check missing values
print(df.isnull().sum())


df['Date/Time'] = pd.to_datetime(df['Date/Time'])

df['Hour'] = df['Date/Time'].dt.hour
df['Day'] = df['Date/Time'].dt.day
df['Month'] = df['Date/Time'].dt.month
df['Weekday'] = df['Date/Time'].dt.day_name()

print(df.head())
```

```python
hourly_trips = df['Hour'].value_counts().sort_index()
print(hourly_trips)
```

```python
import matplotlib.pyplot as plt

hourly_trips = df['Hour'].value_counts().sort_index()

plt.plot(hourly_trips)
plt.xlabel("Hour of the Day")
plt.ylabel("Number of Trips")
plt.title("Trips by Hour")
plt.show()
```

```python
weekday_trips = df['Weekday'].value_counts()
print(weekday_trips)
```

```python
base_trips = df['Base'].value_counts()
print(base_trips)
```

```python
import pandas as pd

# Load multiple months
apr = pd.read_csv("uber-raw-data-apr14.csv")
may = pd.read_csv("uber-raw-data-may14.csv")
jun = pd.read_csv("uber-raw-data-jun14.csv")

# Combine data
df = pd.concat([apr, may, jun])

# Convert Date column
df['Date/Time'] = pd.to_datetime(df['Date/Time'])

# Extract Month
df['Month'] = df['Date/Time'].dt.month

print(df.head())
```

```python
monthly_trips = df['Month'].value_counts().sort_index()
print(monthly_trips)
```

```python
df = df.drop_duplicates()
```

```python
print(df.shape)

df = df.drop_duplicates()

print(df.shape)

df.head()
```

```python
df['Hour'] = df['Date/Time'].dt.hour
df['Weekday'] = df['Date/Time'].dt.day_name()

df.head()
```

```python
df['Date/Time'] = pd.to_datetime(df['Date/Time'])
df.to_csv("cleaned_uber_data.csv", index=False)
```
